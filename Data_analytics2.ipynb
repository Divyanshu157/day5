```python
import openpyxl

def create_excel_file(data):
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    sheet.title = "Food Items"

    header = ["Item", "Place", "Cuisine", "Rate (INR)", "Unit", "Quantity"]
    for col_idx, col in enumerate(header, start=1):
        sheet.cell(row=1, column=col_idx).value = col

    for row, item in enumerate(data):
        for col, value in enumerate(item):
            sheet.cell(row=row+2, column=col).value = value

    workbook.save("food_items.xlsx")

if __name__ == "_main_":
    data = [
        ["Pizza", "Italy", "Italian", 700, 1, 1],
        ["Sushi", "Japan", "Japanese", 350, 1, 1],
        ["Tacos", "Mexico", "Mexican", 140, 1, 1],
        ["Curry", "India", "Indian", 560, 1, 1],
        ["Hamburger", "United States", "American", 350, 1, 1],
        ["Croissant", "France", "French", 140, 1, 1],
        ["Dim Sum", "China", "Chinese", 140, 1, 1],
        ["Paella", "Spain", "Spanish", 1050, 1, 1],
        ["Pad Thai", "Thailand", "Thai", 560, 1, 1],
        ["Moussaka", "Greece", "Greek", 700, 1, 1],
        ["Feijoada", "Brazil", "Brazilian", 700, 1, 1],
        ["Pho", "Vietnam", "Vietnamese", 560, 1, 1],
        ["Kimchi", "South Korea", "Korean", 210, 1, 1],
        ["Kebab", "Turkey", "Turkish", 350, 1, 1],
        ["Asado", "Argentina", "Argentinian", 1050, 1, 1],
        ["Hummus", "Lebanon", "Lebanese", 350, 1, 1],
        ["Nasi Goreng", "Indonesia", "Indonesian", 350, 1, 1],
        ["Couscous", "Morocco", "Moroccan", 560, 1, 1],
        ["Vegemite (spread)", "Australia", "Australian", 280, 1, 1],
        ["Jerk Chicken", "Jamaica", "Jamaican", 560, 1, 1],
        ["Poutine", "Canada", "Canadian", 300, 1, 1],
        ["Goulash", "Hungary", "Hungarian", 500, 1, 1],
    ]

    create_excel_file(data)
```
